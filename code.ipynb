```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Books_Data_Clean.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df[df["Publishing Year"] > 1900]
```

```python
# Checks Na Values
df.isna().sum()
```

```python
# Remove rows with Book name being Na
# inplace = True augments the actual dataset 
df.dropna(subset = "Book Name", inplace = True)
```

```python
# Finding duplicate rows 
# sum() adds up the Falses and/or Trues of the duplicated() functiion. Result is zero if tehre are no duplicates
df.duplicated().sum()
```

```python
# Finding the number of unique values per column
df.nunique()
```

```python
# Histogram of publishing year and it's frequency
plt.hist(df["Publishing Year"])
plt.xlabel("Publishing Year")
plt.ylabel("Frequency")
plt.title("Distribution of Publishing year")
plt.show()
```

```python
#value_counts() finds out how many of the same value in a column
#plot is used to take that data and plots graphs based on what graphs you want
df["genre"].value_counts().plot(kind="bar")
plt.xlabel("Genre")
plt.ylabel("Number of Books")
plt.title("Number of Books in Each Genre")
plt.show()
```

```python
# Groups data by Author and finds the average rating by author
# This data is also shown in descending order, i.e highest average rating to lowest
df.groupby("Author")["Book_average_rating"].mean().sort_values(ascending=False)
```

```python
sns.boxplot(x="genre", y="Book_ratings_count", data=df)
plt.xlabel("Genre")
plt.ylabel("Book Ratings Count")
plt.title("Box Plot of Book Ratings Count for Each Genre")
plt.show()
```

```python
# Plotting a Scatter plot of Sale Price againts Units Sold
# Discovery: The lower the price of the book, the more units it sells
plt.scatter(df["sale price"], df["units sold"])
plt.xlabel("Sale Price")
plt.ylabel("Units Sold")
plt.title("Scatter Plot of Sale Price vs Units Sold")
plt.show()
```

```python
# Finding out what was the most used language in these books.
# Discovery: English was the most used language 
language_counts = df["language_code"].value_counts()
print(language_counts)
```

```python
# Creatinga  pie chart showing the distribution of language codes
# autopct="%1.1f%%" shows the percentages in the chart 
plt.pie(language_counts, labels=language_counts.index, startangle=90, autopct="%1.1f%%")
plt.title("Language Distribution of Books")
plt.show()
```

```python
# Finding the publisher with the highest revenue
# Discovery: It is the Penguin Group (USA) LLC
df.groupby("Publisher ")["publisher revenue"].sum().sort_values(ascending = False)
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].mean().sort_values(ascending=False)
```

```python
df.groupby("Author_Rating")["Book_ratings_count"].max()
```

```python
plt.scatter(df["Book_average_rating"], df['Book_ratings_count'])
plt.xlabel("Book Average Rating")
plt.ylabel("Book Ratings Count")
plt.title("Scatter Plot of Boojk Average Rating vs Book Ratings Count")
plt.show()
```

```python
# Finding the total gross sales per Author in ascending order
total_gross_sales_by_author = df.groupby("Author")["gross sales"].sum().sort_values(ascending=False)
print(total_gross_sales_by_author)
```

```python
# Taking the first 20 rows of the previous solutiona and printed as a Bar Graph
total_gross_sales_by_author.head(20).plot(kind="bar")
plt.xlabel("Author")
plt.ylabel("Total Gross Sales")
plt.title("Total Gross Sales for Each Author")
plt.show()
```

```python
# Showing a box plot of Units Sold per Author Rating.
# Discovery: Those with an Author Ratings of Excellent sell more units compared to other ratings
sns.boxplot(x="Author_Rating", y="units sold", data=df)
plt.xlabel("Author Rating")
plt.ylabel("Units Sold")
plt.title("Box Plots of Units Sold for Each Author Rating")
plt.show()
```

```python
df.groupby("Publishing Year")["units sold"].sum().plot(kind="line", marker="o")
plt.xlabel("Publishing Year")
plt.ylabel("Total Units Sold")
plt.title("Total Units sold Over the Years")
plt.show()
```
